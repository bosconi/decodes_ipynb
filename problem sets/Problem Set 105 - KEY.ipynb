#### 105.02

Rewrite the following code to use a ***procedure*** - that is, a function that that accepts no arguments and returns no values - that captures all the code in the `for fac in faces` loop (but not the loop itself). Name your procedure `subdivide`.

```python
pa,pb,pc,pd = Point(-3,-3),Point(3,-3),Point(3,3),Point(-3,3)
faces = [(Segment(pa,pb),Segment(pb,pc),Segment(pc,pd),Segment(pd,pa))]

#### START YOUR WORK HERE ####

subfaces = []
for fac in faces:
    cen = Point.centroid([seg.ept for seg in fac])
    msubs = [Segment(pt,cen) for pt in [seg.spt for seg in fac]]

    # weave together subdivided segs into four new faces
    subfaces.append( (msubs[0].inverted(),fac[0],msubs[1]) )
    subfaces.append( (msubs[1].inverted(),fac[1],msubs[2]) )
    subfaces.append( (msubs[2].inverted(),fac[2],msubs[3]) )    
    subfaces.append( (msubs[3].inverted(),fac[3],msubs[0]) )
    
#### END YOUR WORK HERE ####    
    
out.put(subfaces)
out.draw()
out.clear()
```

#### 105.06 - Packing and Unpacking Arguments

Packing and unpacking arguments, which can come in both positional and keyword varieties, can be difficult to wrap our heads around. Let's keep things simple. One practical use of the `**kwargs` convention is to use dictionaries to structure the arguments required for the calling of functions and methods.

Working from what we now understand about Decod.es RGons, in the space below, define a function `plot_rgon` that creates a RGon given a dictionary of values, structured as shown in the `things_i_want` variable.
```python
things_i_want = [
    {'type': 'square', 'rad':3},
    {'type': 'triangle','rad':2}, 
    {'type': 'square','rad':1},
    {'type': 'triangle','rad':1},
    {'type': 'square','rad':0.5}
]

#### START YOUR WORK HERE ####

def plot_rgon(**kw):
    if kw['type']=='square': return RGon(4,radius = kw['rad'])
    if kw['type']=='triangle': return RGon(3,radius = kw['rad'])

#### END YOUR WORK HERE ####

for dic in things_i_want:
    rg = plot_rgon(**dic)
    out.put(rg)

out.draw()
out.clear()
```

A List of objects is given, each of which may be a tuple of two Points, and may be a Segment. A `try`-`except` block would be of great value in this case.

```python
some_stuff = [
    Segment(Point.random(Interval(-3,3)),Point.random(Interval(-3,3))),
    (Point.random(Interval(-3,3)),Point.random(Interval(-3,3))),
    Segment(Point.random(Interval(-3,3)),Point.random(Interval(-3,3))),
    (Point.random(Interval(-3,3)),Point.random(Interval(-3,3)))
]

#### START YOUR WORK HERE ####

def middle_point(geom):
    try:
        return geom.eval(0.5)
    except:
        return Point.centroid(geom)

#### END YOUR WORK HERE ####

for stf in some_stuff:
    out.put(middle_point(stf))

out.draw()
out.clear()


```

#### 105.08

A List of random Segments `segs` is given, some of which are longer than the desired maximum length defined by the `max_length` variable. Use the `filter` function to cull the Segments that are too long.

This will require the definition of a function `is_short_enough` which, given a Segment `seg`, returns a boolean value.

```python
max_length = 4.0
cnt_init = 100
segs = [Segment(Point.random(Interval(-3,3)),Point.random(Interval(-3,3))) for n in range(cnt_init)]
cnt_prev = len(segs)

#### START YOUR WORK HERE ####

def is_short_enough(seg):
    return seg.length < max_length

segs = filter(is_short_enough,segs)

#### END YOUR WORK HERE ####

print("{} Segments have been removed".format(cnt_prev- len(segs)))
out.put(segs)
out.draw()
out.clear()
```

#### 105.09

Two Lists of Segments are given, `segs_a` and `segs_b`. 

Author a function `connect_segs` that takes two Segments `seg_a` and `seg_b` as arguments, and returns a Tuple of two Segments: the first connecting a Point 25% along `seg_a` to a Point 75% along `seg_b`, and the second connecting a Point 75% along `seg_a` to a Point 25% along `seg_b`.

Finally, use the `map` function to apply the function `connect_segs` to the two given Lists of Segments, and assign the result to the `cnx_segs`

```python
cnt = 100
segs_a = [Segment(Point(n,-3),Point(n,3)) for n in Interval(-3,3).divide(cnt,True)]
segs_b = [Segment(Point(-3,n),Point(3,n)) for n in Interval(-3,3).divide(cnt,True)]
for seg in segs_a: seg.set_color(0.8)
for seg in segs_b: seg.set_color(0.8)    

#### START YOUR WORK HERE ####

def connect_segs(seg_a,seg_b):
    sa = Segment(seg_a.eval(0.25),seg_b.eval(0.75))
    sb = Segment(seg_a.eval(0.75),seg_b.eval(0.25))
    return(sa,sb)

cnx_segs = map(connect_segs,segs_a,segs_b)

#### END YOUR WORK HERE ####

out.put([segs_a,segs_b])
out.put(cnx_segs)
out.draw()
out.clear()
```

#### 105.10

A Point `attr_pt`, a Vec `t_vec`, and a List of Segments `segs` are given. A procedure at the end of the script iteratively re-defines the list of Segments such that they point in the direction of `attr_pt`. At each step, we intend the `attr_pt` to be nudged to move in the direction and magnitude described by `t_vec`. For this purpose, a yet-to-be-written function `move_attr` is called.

Author the missing function `move_attr`. Since no arguments are passed in the existing code, this function will need to alter the structured object `attr_pt` in a scope outside the function. A special keyword is required for such an action.  

```python
attr_pt = Point(-3,3)
t_vec = Vec(0.5,0)
segs = [Segment(Point(x,0),Point(x,1)) for x in Interval(-4,4).divide(6,True)]


#### START YOUR WORK HERE ####

def move_attr():
    global attr_pt
    attr_pt += t_vec

#### END YOUR WORK HERE ####   
    
for n in range(10):
    move_attr()
    segs = [Segment(seg.spt,Vec(seg.spt,attr_pt).normalized()) for seg in segs]
    
    out.put(segs)
    out.put(attr_pt)

out.draw()
out.clear()
```
